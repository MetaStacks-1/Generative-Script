```python
#!/usr/bin/env python
# coding: utf-8

from PIL import Image 
from IPython.display import display 
import random
import json
import sys
sys.setrecursionlimit(14400)

background = ["Ascension", "Diner Blue", "Diner Red", "Dusk", "Eggshell", "Hot Sauce", "Pink", "Pixel Swamp", "Purple", "Purple Swamp", "Rainbow", "Yoke Sunrise", "Constellation"]
background_weights = [3, 5, 5, 10, 10, 10, 10, 8, 10, 8, 10, 8, 3]

bottom = ["Bacon", "Ham", "Salmon", "Purple Sausage", "Sausage", "Tomato", "Chorizo", "Sludge"]
bottom_weights = [20, 20, 7.5, 2.5, 20, 5, 20, 5]

bun = ["Croissant", "Alien Croissant", "Bageldenza", "Burnt Toast", "Golden Muffin", "Tomato Bagel", "English Muffin", "Multi-Grain Bagel", "Night Muffin", "Rainbow Bagel", "Space Bagel", "Swamp Muffin", "Toad Bagel", "Angry Bagel", "Ape Croissant", "Blue Cat Bagel", "Bones Croissant", "Cyan Bagel", "Ghost Bagel", "Toad Stool Bagel", "Plain Toast", "Toad Croissant"]
bun_weights = [7, 2, .5, 1.5, 1.5, 3, 7, 7, 5, 4, 4, 4, 6, 5, 6, 4.5, 5, 7, 1, 6, 7, 6]

eyes = ["Awake", "3D Glasses", "Bored Ape", "Bored Blue", "Fire Shades", "Noun-Green", "Noun-Teal", "Noun-Violet", "Party Shades", "Morpheus", "Squinted Tomato", "Squinted Sunrise", "Sunrise Eye Roll", "Sus Shades", "Teal Eye Roll"]
eyes_weights = [10, 5, 7.5, 7.5, 3, 3, 3, 3, 3, 7.5, 10, 10, 10, 7.5, 10]

middle = ["Over-Easy", "Scallion-Cream", "Scrambled", "Tomato", "Hashbrowns", "Purple Sausage", "Runny Yoke"]
middle_weights = [10, 25, 25, 5, 25, 2.5, 7.5]

top = ["Avocado", "Vegemite", "Cheddar", "Ketchup", "Pepper-Jack", "Tomato", "Lettuce", "Jam", "Purple Sausage",]
top_weights = [14, 1, 30, 20, 10, 5, 10, 7.5, 2.5]

extras = ["Blue Durag", "Gold Crown", "Plated", "Drive Thru", "Green Durag", "Hot Sauce", "Pink Cap", "Purple Cap", "Short Orange Beanie", "Super Hot Sauce", "Tall Orange Beanie", "None"]
extras_weights = [6, .5, 2.5, 1, 6, 3, 6, 6, 6, 2, 6, 55]


background_files = {
    "Ascension": "ascension",
    "Diner Blue": "dinerblue",
    "Diner Red": "dinerred",
    "Dusk": "dusk",
    "Eggshell": "eggshell",
    "Hot Sauce": "hotsauce",
    "Pink": "pink",
    "Pixel Swamp": "pixelswamp",
    "Purple": "purple",
    "Purple Swamp": "purpleswamp",
    "Rainbow": "rainbow",
    "Yoke Sunrise": "yokesunrise",
    "Constellation": "constellation"
}
bottom_files = {
    "Bacon": "bacon",
    "Ham": "ham",
    "Salmon": "salmon",
    "Purple Sausage": "purplesausage",
    "Sausage": "sausage",
    "Tomato": "tomatobottom",
    "Chorizo": "chorizo",
    "Sludge": "sludge"
}
bun_files = {
    "Croissant": "croissant",
    "Alien Croissant": "aliencroissant",
    "Bageldenza": "bageldenza",
    "Burnt Toast": "burnttoast",
    "Golden Muffin": "goldenmuffin",
    "Tomato Bagel": "tomatobagel", 
    "English Muffin": "englishmuffin", 
    "Multi-Grain Bagel": "multigrainbagel", 
    "Night Muffin": "nightmuffin", 
    "Rainbow Bagel": "rainbowbagel", 
    "Space Bagel": "spacebagel",
    "Swamp Muffin": "swampmuffin",
    "Toad Bagel": "toadbagel",
    "Angry Bagel": "angrybagel",
    "Ape Croissant": "apecroissant",
    "Blue Cat Bagel": "bluecatbagel",
    "Bones Croissant": "bonescroissant",
    "Cyan Bagel": "cyanbagel",
    "Ghost Bagel": "ghostbagel",
    "Toad Stool Bagel": "toadstoolbagel",
    "Plain Toast": "plaintoast",
    "Toad Croissant": "toadcroissant"
}
eyes_files = {
    "Awake": "awake",
    "3D Glasses": "3dglasses",
    "Bored Ape": "boredape",
    "Bored Blue": "boredblue",
    "Fire Shades": "fireshades", 
    "Noun-Green": "noungreen", 
    "Noun-Teal": "nounteal", 
    "Noun-Violet": "nounviolet",
    "Party Shades": "partyshades", 
    "Squinted Sunrise": "squintedsunrise",
    "Morpheus": "morpheus",
    "Squinted Tomato": "squintedtomato",
    "Sunrise Eye Roll": "sunriseeyeroll",
    "Sus Shades": "susshades",
    "Teal Eye Roll": "tealeyeroll"
}
middle_files = {
    "Over-Easy": "overeasy",
    "Scallion-Cream": "scallioncream",
    "Scrambled": "scrambled",
    "Tomato": "tomatomiddle",
    "Purple Sausage": "purplesausagemiddle",
    "Runny Yoke": "runnyyoke",
    "Hashbrowns": "hashbrown"
}
top_files = {
    "Avocado": "avocado",
    "Vegemite": "vegemite",
    "Cheddar": "cheddar",
    "Ketchup": "ketchup",
    "Pepper-Jack": "pepperjack",
    "Tomato": "tomatotop",
    "Lettuce": "lettuce",
    "Purple Sausage": "purplesausagetop",
    "Jam": "jam"
}
extras_files = {
    "Blue Durag": "bluedurag",
    "Gold Crown": "goldcrown",
    "Plated": "plated",
    "Drive Thru": "drivethru",
    "None": "none",
    "Green Durag": "greendurag",
    "Hot Sauce": "hotsauce",
    "Pink Cap": "pinkcap",
    "Purple Cap": "purplecap",
    "Short Orange Beanie": "shortorangebeanie",
    "Super Hot Sauce": "superhotsauce",
    "Tall Orange Beanie": "tallorangebeanie"
}

TOTAL_IMAGES = 6970 # Number of random unique images we want to generate ( 2 x 2 x 2 = 8)

all_images = [] 

def create_new_image():

    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image ["Background"] = random.choices(background, background_weights)[0]
    new_image ["Bottom"] = random.choices(bottom, bottom_weights)[0] 
    new_image ["Bun"] = random.choices(bun, bun_weights)[0]  
    new_image ["Middle"] = random.choices(middle, middle_weights)[0] 
    new_image ["Top"] = random.choices(top, top_weights)[0]
    new_image ["Eyes"] = random.choices(eyes, eyes_weights)[0]
    new_image ["Extras"] = random.choices(extras, extras_weights)[0] 

    if new_image in all_images:
        return create_new_image()
    else:
        return new_image

for i in range(TOTAL_IMAGES): 
        new_trait_image = create_new_image()
        all_images.append(new_trait_image)
        
# Return True If All Images Are Unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))

# Add Token ID to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1

# Print All Images
print(all_images)

# Get Traits Count
background_count = {}
for item in background:
    background_count[item] = 0

bottom_count = {}
for item in bottom:
    bottom_count[item] = 0

bun_count = {}
for item in bun:
    bun_count[item] = 0
    
eyes_count = {}
for item in eyes:
    eyes_count[item] = 0

middle_count = {}
for item in middle:
    middle_count[item] = 0

top_count = {}
for item in top:
    top_count[item] = 0

extras_count = {}
for item in extras:
    extras_count[item] = 0

for image in all_images:
    background_count[image["Background"]] += 1
    bottom_count[image["Bottom"]] += 1
    bun_count[image["Bun"]] += 1
    eyes_count[image["Eyes"]] += 1
    middle_count[image["Middle"]] += 1
    top_count[image["Top"]] += 1
    extras_count[image["Extras"]] += 1

print(background_count)
print(bottom_count)
print(bun_count)
print(eyes_count)
print(middle_count)
print(top_count)
print(extras_count)

METADATA_FILE_NAME = './metadata/all-traits.json'; 
with open(METADATA_FILE_NAME, 'w') as outfile:
    json.dump(all_images, outfile, indent=4)

    
for item in all_images:

    im1 = Image.open(f'./layers/background/{background_files[item["Background"]]}.png').convert("RGBA")
    im2 = Image.open(f'./layers/bottom/{bottom_files[item["Bottom"]]}.png').convert("RGBA")
    im3 = Image.open(f'./layers/middle/{middle_files[item["Middle"]]}.png').convert("RGBA")
    im4 = Image.open(f'./layers/top/{top_files[item["Top"]]}.png').convert("RGBA")
    im5 = Image.open(f'./layers/bun/{bun_files[item["Bun"]]}.png').convert("RGBA")
    im6 = Image.open(f'./layers/extras/{extras_files[item["Extras"]]}.png').convert("RGBA")
    im7 = Image.open(f'./layers/eyes/{eyes_files[item["Eyes"]]}.png').convert("RGBA")
    
    
    com1 = Image.alpha_composite(im1, im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)
    com4 = Image.alpha_composite(com3, im5)
    com5 = Image.alpha_composite(com4, im6)
    com6 = Image.alpha_composite(com5, im7)
    

    #Convert to RGB
    rgb_im = com6.convert("RGB")
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./images/" + file_name)

f = open('./metadata/all-traits.json',) 
data = json.load(f)

IMAGES_BASE_URI = "add_image_base_uri"
PROJECT_NAME = "Brkfst Sndwch"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "description": 'A very rare and carefully crafted Brkfst Sndwch. Here to fuel the metaverse.',
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + '#' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Background", i["Background"]))
    token["attributes"].append(getAttribute("Bottom", i["Bottom"]))
    token["attributes"].append(getAttribute("Bun", i["Bun"]))
    token["attributes"].append(getAttribute("Eyes", i["Eyes"]))
    token["attributes"].append(getAttribute("Middle", i["Middle"]))
    token["attributes"].append(getAttribute("Top", i["Top"]))
    token["attributes"].append(getAttribute("Extras", i["Extras"]))

    with open('./metadata/' + str(token_id) + ".json", 'w') as outfile:
        json.dump(token, outfile, indent=4)
    f.close()
```
